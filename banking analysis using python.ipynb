```python
# IMPORT LIBRARIES
# ============================================

import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# MYSQL CONNECTION
# ============================================

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="@seenu2004",
    database="banking_db"
)

print("Connected Successfully!")
```

```python
# 1. CUSTOMER DISTRIBUTION BY CITY
# ============================================

query1 = """
SELECT city, COUNT(*) AS total_customers
FROM customers
GROUP BY city
"""

df1 = pd.read_sql(query1, conn)

plt.figure(figsize=(8,5))

plt.bar(df1['city'], df1['total_customers'])

plt.title("Customer Distribution by City")
plt.xlabel("City")
plt.ylabel("Total Customers")

plt.show()
```

```python
# 2. ACCOUNT TYPE ANALYSIS
# ============================================

query2 = """
SELECT account_type, COUNT(*) AS total_accounts
FROM accounts
GROUP BY account_type
"""

df2 = pd.read_sql(query2, conn)

plt.figure(figsize=(7,7))

plt.pie(
    df2['total_accounts'],
    labels=df2['account_type'],
    autopct='%1.1f%%'
)

plt.title("Account Type Analysis")

plt.show()
```

```python
# 3. ACCOUNT BALANCE DISTRIBUTION
# ============================================

query3 = """
SELECT balance
FROM accounts
"""

df3 = pd.read_sql(query3, conn)

plt.figure(figsize=(8,5))

plt.hist(df3['balance'], bins=10)

plt.title("Account Balance Distribution")
plt.xlabel("Balance")
plt.ylabel("Frequency")

plt.show()
```

```python
# 4. TRANSACTION TYPE ANALYSIS
# ============================================

query4 = """
SELECT txn_type, COUNT(*) AS total
FROM transactions
GROUP BY txn_type
"""

df4 = pd.read_sql(query4, conn)

plt.figure(figsize=(6,5))

plt.bar(df4['txn_type'], df4['total'])

plt.title("Debit vs Credit Transactions")
plt.xlabel("Transaction Type")
plt.ylabel("Count")

plt.show()
```

```python
# 5. TRANSACTION CHANNEL USAGE
# ============================================

query5 = """
SELECT channel, COUNT(*) AS total
FROM transactions
GROUP BY channel
"""

df5 = pd.read_sql(query5, conn)

plt.figure(figsize=(8,5))

plt.pie(
    df5['total'],
    labels=df5['channel'],
    autopct='%1.1f%%'
)

plt.title("Transaction Channel Usage")

plt.show()
```

```python
# 6. LOAN TYPE DISTRIBUTION
# ============================================

query6 = """
SELECT loan_type, COUNT(*) AS total
FROM loans
GROUP BY loan_type
"""

df6 = pd.read_sql(query6, conn)

plt.figure(figsize=(8,5))

plt.bar(df6['loan_type'], df6['total'])

plt.title("Loan Type Distribution")
plt.xlabel("Loan Type")
plt.ylabel("Count")

plt.xticks(rotation=15)

plt.show()
```

```python
# 7. LOAN STATUS ANALYSIS
# ============================================

query7 = """
SELECT status, COUNT(*) AS total
FROM loans
GROUP BY status
"""

df7 = pd.read_sql(query7, conn)

plt.figure(figsize=(7,5))

plt.bar(df7['status'], df7['total'])

plt.title("Loan Status Analysis")
plt.xlabel("Loan Status")
plt.ylabel("Count")

plt.show()
```

```python
# 8. CUSTOMER AGE ANALYSIS
# ============================================

query8 = """
SELECT
TIMESTAMPDIFF(YEAR, dob, CURDATE()) AS age
FROM customers
"""

df8 = pd.read_sql(query8, conn)

plt.figure(figsize=(8,5))

plt.hist(df8['age'], bins=10)

plt.title("Customer Age Analysis")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# 9. HIGH TRANSACTION CUSTOMERS
# ============================================

query9 = """
SELECT
account_id,
COUNT(*) AS total_transactions
FROM transactions
GROUP BY account_id
ORDER BY total_transactions DESC
LIMIT 10
"""

df9 = pd.read_sql(query9, conn)

plt.figure(figsize=(10,5))

plt.bar(
    df9['account_id'].astype(str),
    df9['total_transactions']
)

plt.title("High Transaction Customers")
plt.xlabel("Account ID")
plt.ylabel("Transactions")

plt.show()
```

```python
# 10. BALANCE VS LOAN AMOUNT COMPARISON
# ============================================

query10 = """
SELECT
a.customer_id,
a.balance,
l.amount AS loan_amount
FROM accounts a
JOIN loans l
ON a.customer_id = l.customer_id
"""

df10 = pd.read_sql(query10, conn)

plt.figure(figsize=(8,5))

plt.scatter(
    df10['balance'],
    df10['loan_amount']
)

plt.title("Balance vs Loan Amount")
plt.xlabel("Account Balance")
plt.ylabel("Loan Amount")

plt.show()
```

```python
# 11. MONTHLY TRANSACTION TREND
# ============================================

query11 = """
SELECT
MONTH(txn_date) AS month,
SUM(amount) AS total_amount
FROM transactions
GROUP BY MONTH(txn_date)
ORDER BY month
"""

df11 = pd.read_sql(query11, conn)

plt.figure(figsize=(10,5))

plt.plot(
    df11['month'],
    df11['total_amount'],
    marker='o'
)

plt.title("Monthly Transaction Trend")
plt.xlabel("Month")
plt.ylabel("Transaction Amount")

plt.grid(True)

plt.show()
```

```python
# 12. BANKING CHANNEL PERFORMANCE
# ============================================

query12 = """
SELECT
channel,
SUM(amount) AS total_amount
FROM transactions
GROUP BY channel
"""

df12 = pd.read_sql(query12, conn)

plt.figure(figsize=(8,5))

plt.bar(
    df12['channel'],
    df12['total_amount']
)

plt.title("Banking Channel Performance")
plt.xlabel("Channel")
plt.ylabel("Transaction Amount")

plt.xticks(rotation=15)

plt.show()
```

```python
# 13. LOAN DEFAULT PREDICTION INSIGHT
# ============================================

query13 = """
SELECT
status,
AVG(amount) AS avg_loan_amount
FROM loans
GROUP BY status
"""

df13 = pd.read_sql(query13, conn)

plt.figure(figsize=(7,5))

plt.bar(
    df13['status'],
    df13['avg_loan_amount']
)

plt.title("Loan Default Prediction Insight")
plt.xlabel("Loan Status")
plt.ylabel("Average Loan Amount")

plt.show()
```

```python
# 14. CUSTOMER FINANCIAL BEHAVIOR ANALYSIS
# ============================================

query14 = """
SELECT
txn_type,
AVG(amount) AS avg_amount
FROM transactions
GROUP BY txn_type
"""

df14 = pd.read_sql(query14, conn)

plt.figure(figsize=(7,5))

plt.bar(
    df14['txn_type'],
    df14['avg_amount']
)

plt.title("Customer Financial Behavior Analysis")
plt.xlabel("Transaction Type")
plt.ylabel("Average Amount")

plt.show()
```

```python
# 15. OVERALL BANKING OPERATIONS DASHBOARD
# ============================================

total_customers = pd.read_sql(
    "SELECT COUNT(*) AS total FROM customers",
    conn
)

total_accounts = pd.read_sql(
    "SELECT COUNT(*) AS total FROM accounts",
    conn
)

total_transactions = pd.read_sql(
    "SELECT COUNT(*) AS total FROM transactions",
    conn
)

total_loans = pd.read_sql(
    "SELECT COUNT(*) AS total FROM loans",
    conn
)

print("\n========== DASHBOARD ==========")

print("Total Customers:",
      total_customers['total'][0])

print("Total Accounts:",
      total_accounts['total'][0])

print("Total Transactions:",
      total_transactions['total'][0])

print("Total Loans:",
      total_loans['total'][0])

# ============================================
# CLOSE CONNECTION
# ============================================

conn.close()

print("\nConnection Closed")
```
